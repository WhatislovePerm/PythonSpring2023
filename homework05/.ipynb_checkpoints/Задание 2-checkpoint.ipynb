## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.
2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
# задание 1.1
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np



x = np.linspace(-1, 1, 250)
y = [(i**2 * np.sin(1 / i**2) + i) for i in x]
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.plot(x, y)
plt.show()
```

```python
# Задание 1.2
x = np.linspace(5, 25, 100)
y = abs(1 - x**2 / (1 + x**2))
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.semilogy(x, y)
```

```python
# Задание 1.3
import scipy as sp
from scipy import special
from scipy.special import logit, expit
x = np.linspace(-5, 5, 100) 
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.plot(x, logit(x), label='logit function', color='gold') 
plt.plot(x, special.ndtri(x), label='Inverse Gaussian cumulative distribution function.')
plt.legend()
```

```python
#Задание 2
x = np.linspace(-6 * np.pi, 6 * np.pi, 250)
y = np.sin(x) / x 
plt.figure(figsize=(20, 8))
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
xtick = np.arange(-6 * np.pi, 6.1 * np.pi, np.pi / 2)
plt.xticks(xtick, list(map(lambda x: x + "Пи", map(str, np.arange(-6, 6.1, 0.5)))))
plt.grid()
plt.plot(x, y)
```

```python
#Задание 3
x = np.linspace(-1 * np.pi, np.pi, 100)
y = np.linspace(-1 * np.pi, np.pi, 100)
XColor, YColor = np.meshgrid(x, y)
z = np.sin(XColor) * np.sin(YColor)
fig, ax = plt.subplots()
normalizeColor = mpl.colors.Normalize(-abs(z).max(), abs(z).max())
p = ax.pcolor(XColor, YColor, z, norm=normalizeColor)
```

```python
#Задание 3.2
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
XColor, YColor = np.meshgrid(x, y)
z = np.e ** (-XColor ** 2 - YColor ** 2)
fig, ax = plt.subplots()
normalizeColor = mpl.colors.Normalize(-abs(z).max(), abs(z).max())
p = ax.pcolor(XColor, YColor, z, norm=normalizeColor)
```
